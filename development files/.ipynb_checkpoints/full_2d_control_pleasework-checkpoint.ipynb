```python
import serial
import time
import threading
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
import pandas as pd

class RobotController:
    def __init__(self, port='COM8', baudrate=9600):
        self.port = port
        self.baudrate = baudrate
        self.ser = None
        self.connected = False
        self.listening = False
        self.listen_thread = None
        
        # Data storage for plotting
        self.telemetry_data = {
            'timestamp': [],
            'x_pos': [],
            'y_pos': [],
            'yaw_angle': [],
            'left_rpm': [],
            'right_rpm': [],
            'state': []
        }
        
        self.connect()
    
    def connect(self):
        """Connect to HC-05 Bluetooth module"""
        try:
            self.ser = serial.Serial(self.port, self.baudrate, timeout=1)
            time.sleep(2)  # Wait for connection to establish
            self.connected = True
            print(f"‚úÖ Connected to HC-05 on {self.port}")
            self.start_listening()
        except serial.SerialException as e:
            print(f"‚ùå Failed to connect to {self.port}: {e}")
            self.connected = False
    
    def disconnect(self):
        """Disconnect from HC-05"""
        self.stop_listening()
        if self.ser and self.ser.is_open:
            self.ser.close()
            self.connected = False
            print("üîå Disconnected from HC-05")
    
    def send_command(self, command):
        """Send command to robot"""
        if not self.connected:
            print("‚ùå Not connected to robot")
            return False
        
        try:
            self.ser.write((command + '\n').encode())
            print(f"üì§ Sent: {command}")
            return True
        except Exception as e:
            print(f"‚ùå Failed to send command: {e}")
            return False
    
    def start_listening(self):
        """Start listening for incoming data in a separate thread"""
        if self.listening:
            return
        
        self.listening = True
        self.listen_thread = threading.Thread(target=self._listen_loop, daemon=True)
        self.listen_thread.start()
        print("üëÇ Started listening for robot data...")
    
    def stop_listening(self):
        """Stop listening for incoming data"""
        self.listening = False
        if self.listen_thread:
            self.listen_thread.join(timeout=1)
        print("üîá Stopped listening")
    
    def _listen_loop(self):
        """Main listening loop (runs in separate thread)"""
        while self.listening and self.connected:
            try:
                if self.ser.in_waiting > 0:
                    line = self.ser.readline().decode('utf-8').strip()
                    if line:
                        self._process_received_data(line)
            except Exception as e:
                print(f"‚ùå Error reading data: {e}")
                time.sleep(0.1)
            time.sleep(0.01)  # Small delay to prevent excessive CPU usage
    
    def _process_received_data(self, data):
        """Process and parse received data from robot"""
        print(f"üì• Received: {data}")
        
        # Parse telemetry data if it matches expected format
        # Expected format: "State:X,X:Y,Y:Z,Yaw:A,L_RPM:B,R_RPM:C"
        if data.startswith("State:"):
            try:
                self._parse_telemetry(data)
            except Exception as e:
                print(f"‚ö†Ô∏è Failed to parse telemetry: {e}")
    
    def _parse_telemetry(self, data):
        """Parse telemetry data and store for plotting"""
        # Parse the telemetry string
        parts = data.split(',')
        
        state = parts[0].split(':')[1]
        x_pos = float(parts[1].split(':')[1])
        y_pos = float(parts[2].split(':')[1])
        yaw = float(parts[3].split(':')[1])
        left_rpm = float(parts[4].split(':')[1])
        right_rpm = float(parts[5].split(':')[1])
        
        # Store data with timestamp
        current_time = datetime.now()
        self.telemetry_data['timestamp'].append(current_time)
        self.telemetry_data['x_pos'].append(x_pos)
        self.telemetry_data['y_pos'].append(y_pos)
        self.telemetry_data['yaw_angle'].append(yaw)
        self.telemetry_data['left_rpm'].append(left_rpm)
        self.telemetry_data['right_rpm'].append(right_rpm)
        self.telemetry_data['state'].append(state)
        
        # Keep only last 100 data points to prevent memory issues
        if len(self.telemetry_data['timestamp']) > 100:
            for key in self.telemetry_data:
                self.telemetry_data[key] = self.telemetry_data[key][-100:]
    
    def go_to(self, x, y):
        """Send robot to specific coordinates"""
        command = f"{x},{y}"
        return self.send_command(command)
    
    def get_status(self):
        """Get current robot status"""
        if not self.telemetry_data['timestamp']:
            return "No telemetry data available"
        
        latest_idx = -1
        return {
            'position': (self.telemetry_data['x_pos'][latest_idx], 
                        self.telemetry_data['y_pos'][latest_idx]),
            'yaw': self.telemetry_data['yaw_angle'][latest_idx],
            'state': self.telemetry_data['state'][latest_idx],
            'rpm': (self.telemetry_data['left_rpm'][latest_idx], 
                   self.telemetry_data['right_rpm'][latest_idx])
        }
    
    def plot_trajectory(self):
        """Plot robot's trajectory and telemetry"""
        if not self.telemetry_data['timestamp']:
            print("No data to plot")
            return
        
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 8))
        
        # Plot 1: 2D Trajectory
        ax1.plot(self.telemetry_data['x_pos'], self.telemetry_data['y_pos'], 'b-', linewidth=2)
        ax1.scatter(self.telemetry_data['x_pos'][0], self.telemetry_data['y_pos'][0], 
                   color='green', s=100, label='Start', marker='o')
        ax1.scatter(self.telemetry_data['x_pos'][-1], self.telemetry_data['y_pos'][-1], 
                   color='red', s=100, label='Current', marker='x')
        ax1.set_xlabel('X Position (mm)')
        ax1.set_ylabel('Y Position (mm)')
        ax1.set_title('Robot Trajectory')
        ax1.grid(True)
        ax1.legend()
        ax1.axis('equal')
        
        # Plot 2: Yaw Angle over time
        times = [(t - self.telemetry_data['timestamp'][0]).total_seconds() 
                for t in self.telemetry_data['timestamp']]
        ax2.plot(times, self.telemetry_data['yaw_angle'], 'r-', linewidth=2)
        ax2.set_xlabel('Time (seconds)')
        ax2.set_ylabel('Yaw Angle (degrees)')
        ax2.set_title('Orientation over Time')
        ax2.grid(True)
        
        # Plot 3: Motor RPM
        ax3.plot(times, self.telemetry_data['left_rpm'], 'b-', label='Left Motor', linewidth=2)
        ax3.plot(times, self.telemetry_data['right_rpm'], 'r-', label='Right Motor', linewidth=2)
        ax3.set_xlabel('Time (seconds)')
        ax3.set_ylabel('RPM')
        ax3.set_title('Motor Speeds')
        ax3.grid(True)
        ax3.legend()
        
        # Plot 4: Position vs Time
        ax4.plot(times, self.telemetry_data['x_pos'], 'b-', label='X Position', linewidth=2)
        ax4.plot(times, self.telemetry_data['y_pos'], 'r-', label='Y Position', linewidth=2)
        ax4.set_xlabel('Time (seconds)')
        ax4.set_ylabel('Position (mm)')
        ax4.set_title('Position over Time')
        ax4.grid(True)
        ax4.legend()
        
        plt.tight_layout()
        plt.show()
    
    def save_data(self, filename=None):
        """Save telemetry data to CSV file"""
        if not self.telemetry_data['timestamp']:
            print("No data to save")
            return
        
        if filename is None:
            filename = f"robot_data_{datetime.now().strftime('%Y%m%d_%H%M%S')}.csv"
        
        df = pd.DataFrame(self.telemetry_data)
        df.to_csv(filename, index=False)
        print(f"üíæ Data saved to {filename}")
    
    def clear_data(self):
        """Clear all stored telemetry data"""
        for key in self.telemetry_data:
            self.telemetry_data[key] = []
        print("üóëÔ∏è Telemetry data cleared")

# Usage Examples and Helper Functions
def demo_commands(robot):
    """Demonstrate various robot commands"""
    print("\nü§ñ Robot Demo Commands:")
    print("1. Square pattern")
    print("2. Circle approximation")
    print("3. Return to origin")
    
    choice = input("Enter choice (1-3): ")
    
    if choice == '1':
        # Square pattern
        print("Executing square pattern...")
        robot.go_to(100, 0)
        time.sleep(5)
        robot.go_to(100, 100)
        time.sleep(5)
        robot.go_to(0, 100)
        time.sleep(5)
        robot.go_to(0, 0)
        
    elif choice == '2':
        # Circle approximation (octagon)
        print("Executing circle approximation...")
        radius = 100
        for i in range(8):
            angle = i * 2 * np.pi / 8
            x = radius * np.cos(angle)
            y = radius * np.sin(angle)
            robot.go_to(x, y)
            time.sleep(3)
        robot.go_to(0, 0)
        
    elif choice == '3':
        print("Returning to origin...")
        robot.go_to(0, 0)

# Main execution
if __name__ == "__main__":
    # Create robot controller instance
    robot = RobotController(port='COM8', baudrate=9600)
    
    if robot.connected:
        print("\nüéÆ Robot Control Interface Ready!")
        print("Commands:")
        print("  robot.go_to(x, y) - Move to coordinates")
        print("  robot.get_status() - Get current status")
        print("  robot.plot_trajectory() - Plot movement data")
        print("  robot.save_data() - Save data to CSV")
        print("  robot.clear_data() - Clear stored data")
        print("  demo_commands(robot) - Run demo patterns")
        print("  robot.disconnect() - Disconnect")
        
        # Example usage
        print("\nüìç Example: Moving to (50, 50)")
        robot.go_to(50, 50)
        
        # Wait a bit then show status
        time.sleep(2)
        print(f"\nüìä Current Status: {robot.get_status()}")
    
    else:
        print("Failed to connect to robot. Check COM port and HC-05 connection.")

# Jupyter Notebook Helper Functions
def quick_connect():
    """Quick connect function for Jupyter"""
    return RobotController(port='COM8', baudrate=9600)

def interactive_control(robot):
    """Interactive control for Jupyter cells"""
    print("Interactive Robot Control")
    print("Usage in Jupyter cells:")
    print("robot.go_to(100, 0)  # Move to (100, 0)")
    print("robot.go_to(0, 100)  # Move to (0, 100)")
    print("robot.plot_trajectory()  # Show plots")
    return robot
```
